```python
import numpy as np


def f(x, y):
    return x**4 + x**2 * y**2 - y**2 + y**4 + 6


def gradient_f(x, y):
    df_dx = 4 * x**3 + 2 * x * y**2
    df_dy = 2 * x**2 * y - 2 * y + 4 * y**3
    return np.array([df_dx, df_dy])


def gradient_descent(initial_point, learning_rate, num_iterations):
    current_point = np.array(initial_point, dtype=float)

    for _ in range(num_iterations):
        gradient = gradient_f(current_point[0], current_point[1])
        current_point -= learning_rate * gradient

    return current_point, f(current_point[0], current_point[1])

initial_point = [1, 1]
learning_rate = 0.02
num_iterations = 500


minima_point, minima_value = gradient_descent(initial_point, learning_rate, num_iterations)

print("Minima Point:", minima_point)
print("Minima Value:", minima_value)
```
